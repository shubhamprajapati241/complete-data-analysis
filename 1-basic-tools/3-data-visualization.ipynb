# Matplotlib
- Powerful plotting library for Python that enables the creation of static, animated and interactive visualization.

```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [1,4,6,16,29]

# create the line plot
plt.plot(x,y)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Basic line plot")
plt.show()
```

```python
# creating a customized line plot
plt.plot(x,y,color='red', linestyle='-.',marker='o', linewidth=3, markersize=10)
plt.grid(True)
```

```python
# Multiple plots
x= [1,2,3,4,5]
y1 = [1,4,9,16,25]
y2= [1,2,3,4,5]

plt.figure(figsize=(9,5))
plt.subplot(2,2,1)
plt.plot(x,y1,color="green")
plt.title("Plot1")

plt.subplot(2,2,2)
plt.plot(y1,x, color="red")
plt.title("Plot2")

plt.subplot(2,2,3)
plt.plot(y1, y2, color="blue")
plt.title("Plot3")

plt.subplot(2,2,4)
plt.plot(y2,y1,color="orange")
plt.title("Plot4")
```

```python
# Bar plot 

categories = ['A','B', 'C','D','E']
value=[5,6,7,8,9]

plt.bar(categories, value)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Bar plot")
plt.show()
```

# Histogram
- represents the distribution of a dataset. 
- divide the data into bins and count the number of data points in each bin

```python
data = [1,2,2,3,4,4,5,5,6,6,6,7,7,7,7]
plt.hist(data, bins=5, color='orange', edgecolor='black')
```

```python
# creating a scatter plot
x=[1,2,3,4,5]
y=[2,3,4,5,6]

plt.scatter(x,y,color='blue', marker='x')
```

```python
# Creating a pipe chart
labels = ['A','B','C','D']
sizes = [30, 20, 40, 10]
colors = ['gold', 'yellow', 'lightcoral', 'lightskyblue']
explode = (0.1,0.2,0.05,0.3)

plt.pie(sizes, labels=labels, colors=colors, explode=explode, autopct="%1.1f%%", shadow=True)
```

# Seaborn 
- Based on Matplotlib
- Provide a high-level interface for drawing attactive and informative statistical graphics
- Helps in creating complex visualization 

```python
!pip install seaborn
```

```python
import seaborn as sns
```

```python
# basic plotting with seaborn 
tips = sns.load_dataset('tips')
tips
```

```python
# create a scatter plot
import matplotlib.pyplot as plt
sns.scatterplot(x="total_bill", y="tip", data= tips)
plt.title("Scatter plot of Total Bill vs Tips")
plt.show()
```

```python
# Line plot
sns.lineplot(x="size", y="total_bill", data=tips)
plt.title("Line Plot : Size * Total Bill")
```

```python
# categorical plot 

# 1. Bar plot
sns.barplot(x='day', y='total_bill',data=tips)
plt.title("Bar plot : total_bill * day")
plt.show()
```

```python
# Box plot
sns.boxplot(x="day", y="total_bill", data=tips)
plt.title("Box plot : total_bill vs day")
plt.show()
```

```python
# violin plot
sns.violinplot(x='day', y='total_bill', data=tips)
```

```python
# Histogram
sns.histplot(tips['total_bill'], bins=10,kde=True)
# kde - kernal density estimator
```

```python
# KDE plot
sns.kdeplot(tips['total_bill'], fill=True)
```

```python
# pairplot -> relationship between all the variable to difference variables -> best for numerical variable
sns.pairplot(tips)
```

```python
# Heatmap -> with numberical variable
corr = tips[['total_bill', 'tip', 'size']].corr()
corr
sns.heatmap(corr, annot=True, cmap="coolwarm")
```
